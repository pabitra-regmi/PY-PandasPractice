```python
import pandas as pd

file_name = "Panchami Balancesheet.xlsx"
df = pd.read_excel(file_name, engine='openpyxl')

def find_text_errors(df):
    errors = {}
    for column in df.columns:
        numeric_col = pd.to_numeric(df[column], errors='coerce')
        error_rows = df[numeric_col.isna() & df[column].notna()]
        if not error_rows.empty:
            errors[column] = error_rows
    return errors

errors_found = find_text_errors(df)

if errors_found:
    print(" Text Errors Found in Excel:")
    for col, rows in errors_found.items():
        print(f"\nColumn: {col}")
        print(rows)
else:
    print("No text errors found in numeric columns!")
```

```python
import pandas as pd

file_name = "Panchami Balancesheet.xlsx"  
df = pd.read_excel(file_name, engine='openpyxl')

df.columns = df.columns.str.strip()

print("Columns in Excel:", df.columns.tolist())

duplicate_rows = df[df.duplicated()]
if not duplicate_rows.empty:
    print("Duplicate rows found:")
    print(duplicate_rows)
else:
    print("No duplicate rows found.")

column_name = 'ledgerHead'  
if column_name in df.columns:
    duplicate_in_column = df[df[column_name].duplicated()]
    if not duplicate_in_column.empty:
        print(f"\nDuplicates in column '{column_name}':")
        print(duplicate_in_column[[column_name]])
    else:
        print(f"No duplicates found in column '{column_name}'.")
else:
    print(f"Column '{column_name}' not found in the Excel file.")
```
